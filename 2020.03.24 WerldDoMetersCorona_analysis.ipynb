```python
import requests
import pandas as pd
from datetime import date, timedelta 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#importing the corona aatal van de landen
#df = pd.read_csv('output/corona.csv',index_col=0)
df = pd.read_csv('output/corona_dailyNewCases.csv',index_col=0)
df
```

```python
df1 = df
df1
```

```python
populations=[60480000,327200000,46660000,82790000,66890000,8560000,66440000,17180000]
i=0
for column in df.drop('date', axis=1):
    df1[column]= (df1[column].astype(int))/populations[i]
    i+=1
df1
```

```python
df1.to_csv('output/corona%.csv')
```

```python
#plt.style.use('fivethirtyeight')
plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(1080/my_dpi, 200/my_dpi), dpi=my_dpi)
 
# plot
#plt.plot( 'date', 'Netherlands', data=df)
plt.plot( 'date', 'Netherlands', data=df1)
plt.show()
```

```python
#plt.style.use('fivethirtyeight')
def plott(column):
    #plt.style.use('fivethirtyeight')
    plt.style.use('seaborn-darkgrid')
    my_dpi=96
    plt.figure(figsize=(16,2), dpi=my_dpi)
    # plot    
    plt.plot( 'date', column, data=df1,)
    plt.plot( 'date', 'Netherlands', data=df1)
    print(column + ' & Netherlands --- Comparison of Daily Cases by Population')
    plt.show()
```

```python
for column in df1.drop('date', axis=1):
    plott(column)
```

```python
# Draw Plot
plt.figure(figsize=(16,5), dpi= 80)
for column in df1.drop('date', axis=1):
    sns.kdeplot(df1[column], shade=False,alpha=.9)

# Decoration
plt.title('Density Plot of City Mileage by n_Cylinders', fontsize=22)
#plt.legend()
plt.show()
```

```python
# Draw Plot
def Density(column):
    plt.figure(figsize=(16,2), dpi= 80)
    sns.kdeplot(df1[column], shade=False,alpha=.9)
    sns.kdeplot(df1['Netherlands'], shade=False,alpha=.9)
    # Decoration
    plt.title('Density Plot of ' + str(column), fontsize=12)
    #plt.legend()
    plt.show()
    
```

```python
for column in df1.drop('date', axis=1):
    Density(column)
```
