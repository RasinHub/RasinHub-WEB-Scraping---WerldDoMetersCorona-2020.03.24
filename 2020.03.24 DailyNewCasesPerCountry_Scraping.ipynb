```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import date, timedelta 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url='https://www.worldometers.info/coronavirus/#countries'
r=requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
#class="wikitable plainrowheaders sortable jquery-tablesorter"
table=soup.find('table',{'id':'main_table_countries_yesterday'})
rows=table.find_all('tr')[1:]
rows[1]
```

```python
italy=0
usa=0
spain=0
germany=0
france=0
switzerland=0
uk=0
netherlands=0
aantalen=[]
yesterday = date.today() - timedelta(days=1)
yesterday=yesterday.strftime("%m/%d/%Y")
aantalen.append(str(yesterday))
#country_names=['Italy','USA','Spain','Germany','France','Switzerland','UK','Netherlands']
for row in rows:   
    country=row.find_all('td')[0].text.strip() # The column of Country Names 
    #print(country)
    aantal=row.find_all('td')[2].text.strip() #the Column of Daily New Cases 
    #print(aantal)
    #!!!!!! IMPORTANT: Aşağıdaki Sıralamayı değiştirmeyin
    if country == 'Italy':
        italy = aantal.replace(",", "")
        italy = italy.replace("+", "")
    if country == 'USA':
        usa = aantal.replace(",", "")
        usa = usa.replace("+", "")
    if country == 'Spain':
        spain= aantal.replace(",", "")
        spain = spain.replace("+", "")
    if country == 'Germany':
        germany=aantal.replace(",", "")
        germany = germany.replace("+", "")
    if country == 'France':
        france = aantal.replace(",", "")
        france = france.replace("+", "")
    if country == 'Switzerland':
        switzerland = aantal.replace(",", "")
        switzerland = switzerland.replace("+", "")
    if country == 'UK':
        uk = aantal.replace(",", "")
        uk = uk.replace("+", "")
    if country == 'Netherlands':
        netherlands = aantal.replace(",", "")
        netherlands = netherlands.replace("+", "")
aantalen.append(italy)
aantalen.append(usa)
aantalen.append(spain)
aantalen.append(germany)
aantalen.append(france)
aantalen.append(switzerland)
aantalen.append(uk)
aantalen.append(netherlands)


#print(countries)
print(aantalen)
```

```python
#importing the corona aatal van de landen
df = pd.read_csv('output/corona_dailyNewCases.csv',index_col=0)
df = df.drop_duplicates()
df
```

```python
aantalen[0] in str(df['date'])
```

```python
if aantalen[0] in str(df['date']):
    print('data is already updated!')
else:    
    df.loc[len(df)] = aantalen
    df.to_csv('output/corona_dailyNewCases.csv')
    print('data is saved')
    
```
